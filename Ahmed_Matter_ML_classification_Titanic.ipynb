```python
#data cleaning

columns_to_remove_train = ['PassengerId', 'Name', 'Ticket', 'Cabin']
columns_to_remove_test = ['Name', 'Ticket', 'Cabin']

age_median = train_df['Age'].median()
fare_median = train_df['Fare'].median()

train_df = train_df.drop(columns=columns_to_remove_train, axis=1)
test_df = test_df.drop(columns=columns_to_remove_test, axis=1)


train_df['Age'] = train_df['Age'].fillna(age_median)
test_df['Age'] = test_df['Age'].fillna(age_median)

train_df['Fare'] = train_df['Fare'].fillna(fare_median)
test_df['Fare'] = test_df['Fare'].fillna(fare_median)

train_df.dropna(subset=['Embarked'], inplace=True)
test_df.dropna(subset=['Embarked'], inplace=True)
```

**Second : Encoding & Scaling**

```python
#Encoding

#Encode Sex Col

gender_mapping = {'male': 0, 'female': 1}


train_df['Sex'] = train_df['Sex'].map(gender_mapping)
test_df['Sex'] = test_df['Sex'].map(gender_mapping)

encoder = OneHotEncoder(sparse_output=False)
#Encode embarked Col
embarked_encoded_train = encoder.fit_transform(train_df[['Embarked']])
embarked_encoded_test = encoder.transform(test_df[['Embarked']])


encoded_columns = encoder.get_feature_names_out(['Embarked'])

train_embarked_df = pd.DataFrame(embarked_encoded_train, columns=encoded_columns, index=train_df.index)
test_embarked_df = pd.DataFrame(embarked_encoded_test, columns=encoded_columns, index=test_df.index)


train_df = pd.concat([train_df.drop(columns=['Embarked']), train_embarked_df], axis=1)
test_df = pd.concat([test_df.drop(columns=['Embarked']), test_embarked_df], axis=1)

#Encode Pclass Col
pclass_encoded_train = encoder.fit_transform(train_df[['Pclass']])
pclass_encoded_test = encoder.transform(test_df[['Pclass']])

encoded_columns = encoder.get_feature_names_out(['Pclass'])

train_pclass_df = pd.DataFrame(pclass_encoded_train, columns=encoded_columns, index=train_df.index)
test_pclass_df = pd.DataFrame(pclass_encoded_test, columns=encoded_columns, index=test_df.index)

train_df = pd.concat([train_df.drop(columns=['Pclass']), train_pclass_df], axis=1)
test_df = pd.concat([test_df.drop(columns=['Pclass']), test_pclass_df], axis=1)


#Feature Engineering by adding 2cols familysize and is alone

train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
test_df['FamilySize'] = test_df['SibSp'] + test_df['Parch'] + 1


train_df['IsAlone'] = (train_df['FamilySize'] == 1).astype(int)
test_df['IsAlone'] = (test_df['FamilySize'] == 1).astype(int)



#Drop columns SibSp ,Parch

columns_to_drop = ['SibSp', 'Parch']

train_df = train_df.drop(columns=columns_to_drop, axis=1)
test_df = test_df.drop(columns=columns_to_drop, axis=1)
```

```python
#Scaling

scaler = StandardScaler()

numerical_cols = ['Age', 'Fare']

train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])

test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])
```
